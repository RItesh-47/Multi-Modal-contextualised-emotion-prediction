```python
videoLabel={}
videoSpeaker={}
videoID={}
videoText={}
videoAudio={}
videoSentence={}

# for col in ['emotion', 'text_embedding', 'audio_embedding', 'roman_transcript', 'Speaker']:
#     for val1, val2 in zip(df.loc[:, "Label"], df.loc[:, col]):
#         if col=="emotion":
#             videoLabel[df.loc[val1,'Label'].astype('str')] = [(df.loc[val2, col]).astype('str')]
#         elif col=="roman_transcript":
#             videoSentence[df.loc[val1,'Label'].astype('str')] = [(df.loc[val2, col]).astype('str')]
#         elif col=="Speaker":
#             videoLabel[df.loc[val1,'Label'].astype('str')] = [(df.loc[val2, col]).astype('str')]
#         elif col=="text_embedding":
#             videoText[df.loc[val1,'Label'].astype('str')] = [np.array(df.loc[val2, col])]
#         elif col=="videoAudio":
#             videoText[df.loc[val1,'Label'].astype('str')] = [np.array(df.loc[val2, col])]
        
for col in ['emotion', 'text_embedding', 'audio_embedding', 'roman_transcript', 'Speaker']:
    for idx, row in df.iterrows():
        label_str = str(row['Label'])
        if col == "emotion":
            videoLabel[label_str] = [int(row[col])]
        elif col == "roman_transcript":
            videoSentence[label_str] = [str(row[col])]
        elif col == "Speaker":
            videoSpeaker[label_str] = [str(row[col])]
        elif col == "text_embedding":
            videoText[label_str] = [np.array(row[col])]
        elif col == "audio_embedding":
            videoAudio[label_str] = [np.array(row[col])]
```

```python
from sklearn.model_selection import train_test_split

# Assuming df is your DataFrame and 'Label' is the column with labels
labels = df['Label'].unique()
labels = [str(label) for label in labels]

# Split the labels into train (80%) and test (20%) sets
train_labels, test_labels = train_test_split(labels, test_size=0.2, random_state=42)

# Convert to sets for easy lookup
train_set = set(train_labels)
test_set = set(test_labels)

print(f"Total labels: {len(labels)}")
print(f"Train labels: {len(train_set)}")
print(f"Test labels: {len(test_set)}")
```

```python
videoVisual={}
for idx, row in df.iterrows():
    label_str = str(row['Label'])
    videoVisual[label_str] = [np.array(0)]
```

```python
import pickle as pkl
data_to_save = (videoID, videoSpeaker, videoLabel, videoText, videoAudio, videoVisual,
videoSentence, train_set, test_set)
with open('./stashfin_dialoguernn_v1.pkl', 'wb') as f:
    pkl.dump(data_to_save, f, protocol=pkl.HIGHEST_PROTOCOL)
```
